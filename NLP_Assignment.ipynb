# Countervectorizer

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression, SGDClassifier, PassiveAggressiveClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

def countervector():
    # Load dataset
    df = pd.read_csv('whatsapp_chat.csv')
    dataset = df.drop(columns=["Unnamed: 0"], errors="ignore")   
    X = dataset['Message']
    y = dataset['Flirt_Label']
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=53)

    # CountVectorizer
    count_vectorizer = CountVectorizer(stop_words='english')
    count_train = count_vectorizer.fit_transform(X_train)
    count_test = count_vectorizer.transform(X_test)

    # Define classifiers
    classifiers = {
        "MultinomialNB": MultinomialNB(),
        "BernoulliNB": BernoulliNB(),
        "LogisticRegression": LogisticRegression(max_iter=2000),
        "LinearSVC": LinearSVC(),
        "SGDClassifier": SGDClassifier(),
        "PassiveAggressive": PassiveAggressiveClassifier(),
        "RandomForest": RandomForestClassifier()
    }

    # Store results
    results = []

    # Train & evaluate each classifier
    for name, clf in classifiers.items():
        clf.fit(count_train, y_train)
        pred = clf.predict(count_test)
        acc = metrics.accuracy_score(y_test, pred)
        results.append((name, acc))
        print(f"{name} Accuracy: {acc:.4f}")

    # Sort results by accuracy
    results.sort(key=lambda x: x[1], reverse=True)

    print("\n=== Accuracy Ranking ===")
    for name, acc in results:
        print(f"{name}: {acc:.4f}")

# Run the function
countervector()
```

# TfidfVectorizer

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.linear_model import LogisticRegression, SGDClassifier, PassiveAggressiveClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

def tfidfvectorizer():
    # Load dataset
    df = pd.read_csv('whatsapp_chat.csv')
    dataset = df.drop(columns=["Unnamed: 0"], errors="ignore")   
    X = dataset['Message']
    y = dataset['Flirt_Label']
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=53)

    # TfidfVectorizer
    tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
    tfidf_train = tfidf_vectorizer.fit_transform(X_train)
    tfidf_test = tfidf_vectorizer.transform(X_test)

    # Define classifiers
    classifiers = {
        "MultinomialNB": MultinomialNB(),
        "BernoulliNB": BernoulliNB(),
        "LogisticRegression": LogisticRegression(max_iter=2000),
        "LinearSVC": LinearSVC(),
        "SGDClassifier": SGDClassifier(),
        "PassiveAggressive": PassiveAggressiveClassifier(),
        "RandomForest": RandomForestClassifier()
    }

    # Store results
    results = []

    # Train & evaluate each classifier
    for name, clf in classifiers.items():
        clf.fit(tfidf_train, y_train)
        pred = clf.predict(tfidf_test)
        acc = metrics.accuracy_score(y_test, pred)
        results.append((name, acc))
        print(f"{name} Accuracy: {acc:.4f}")

    # Sort results by accuracy
    results.sort(key=lambda x: x[1], reverse=True)

    print("\n=== Accuracy Ranking ===")
    for name, acc in results:
        print(f"{name}: {acc:.4f}")

# Run the function
tfidfvectorizer()
```

# Final Model

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

def final_model():
    # Load dataset
    df = pd.read_csv('whatsapp_chat.csv')
    dataset = df.drop(columns=["Unnamed: 0"], errors="ignore")
    X = dataset['Message']
    y = dataset['Flirt_Label']
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.33, random_state=53
    )

    # CountVectorizer
    count_vectorizer = CountVectorizer(stop_words='english')
    count_train = count_vectorizer.fit_transform(X_train)
    count_test = count_vectorizer.transform(X_test)

   
    clf = MultinomialNB()
    clf.fit(count_train, y_train)

    # Evaluate
    pred = clf.predict(count_test)
    acc = metrics.accuracy_score(y_test, pred)
    print(f"Final Model (MultinomialNB) Accuracy: {acc:.4f}")

    # === Example prediction on new messages ===
    new_messages = [
        "Hey you look amazing today ",
        "Meeting at 5 PM in office",
        "Can we go out for dinner?"
    ]
    new_transformed = count_vectorizer.transform(new_messages)
    predictions = clf.predict(new_transformed)

    for msg, label in zip(new_messages, predictions):
        print(f"Message: '{msg}' --> Prediction: {label}")

# Run final model
final_model()
```

# Check flirt messages

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB


df = pd.read_csv("whatsapp_chat.csv")

# Ensure Message is string
X = df["Message"].astype(str)       # Input features (messages)
y = df["Flirt_Label"]               # Target (0 = Not Flirt, 1 = Flirt)


vectorizer = CountVectorizer(stop_words="english")
X_vec = vectorizer.fit_transform(X)


model = MultinomialNB()
model.fit(X_vec, y)

df["Flirt_Prediction"] = model.predict(X_vec)   # Predicted label

# Prediction probabilities
probs = model.predict_proba(X_vec)
df["Not_Flirt_Prob"] = probs[:, 0]
df["Flirt_Prob"] = probs[:, 1]


print(df.head(10))   # Show first 10 rows
df.to_csv("flirt_predictions.csv", index=False)
print("\n Results saved to flirt_predictions.csv")
```
